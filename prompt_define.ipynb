```python
cp1 = ChatPromptTemplate.from_messages(
                                [('system','you are a {role}'),
                                 ('human','{question}')
                                ],

                )
```

```python
cp1.invoke({"role":"marketing analyst","question":"what is ab testing"})
```

```python
cp1.format(role = "marketing analyst",question = "what is ab testing")
```

```python
ChatPromptTemplate.from_messages(
                                [
                                 ('system','you are a {role}'),
                                 ('human','{question}')
                                ],
                                partial_variables={'role':'Data Scientist'}
                )
```

```python
prompt = ChatPromptTemplate(
    messages=[
        ("system", "You are a {role}."),
        ("human", "{question}")
    ],
    partial_variables={"role": "Marketing Analyst"}
)
```

```python
prompt.format_messages(question='who is messi')
```

```python
prompt.invoke({'question':'Who is messi'})
```
